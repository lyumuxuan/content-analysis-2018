```python
# convert string representation back to list
string_represent_variables = ['tokenized_abstract', 'tokenized_title', 
                              'normalized_abstract', 'normalized_title', 
                              'psyc_check']
for variable in string_represent_variables:
    pubmed[variable] = pubmed[variable].apply(ast.literal_eval)
pubmed
```

```python
sample = pubmed[-10000:].copy()
```

```python
countsDict = {}
for word in sample['normalized_abstract'].sum():
    if word in countsDict:
        countsDict[word] += 1
    else:
        countsDict[word] = 1
word_counts = sorted(countsDict.items(), key = lambda x : x[1], reverse = True)
word_counts[:100]
```

```python
pmBigrams = nltk.collocations.BigramCollocationFinder.from_words(sample['normalized_abstract'].sum())
print("There are {} bigrams in the finder".format(pmBigrams.N))
```

```python
bigram_measures = nltk.collocations.BigramAssocMeasures()
pmBigrams.score_ngrams(bigram_measures.likelihood_ratio)[:100]
# other options include student_t, chi_sq, likelihood_ratio, pmi
```

```python
trigram_measures = nltk.collocations.TrigramAssocMeasures()
pmTrigrams = nltk.collocations.TrigramCollocationFinder.from_words(sample['normalized_abstract'].sum())
pmTrigrams.score_ngrams(trigram_measures.likelihood_ratio)[:100]
```
