```python
#word freq count by documents
newcount = {}
#for word in new['tokens'].sum():
for i in range(2689):
    countsDict = {}
    for word in pubmed_80to89['normalized_abstract'][i]:
        if word in countsDict:
            countsDict[word] = 1
        else:
            countsDict[word] = 1
            if word not in newcount.keys():
                newcount[word] = 0
            newcount[word] += 1
    
word_counts = sorted(newcount.items(), key = lambda x : x[1], reverse = True)
```
