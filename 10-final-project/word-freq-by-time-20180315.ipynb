```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from collections import Counter
```

```python
# read data from pickle (15 seconds)
# pubmed.to_pickle("../data/combined_data.pkl")
pubmed_all = pd.read_pickle("../data/date_cleaned_data.pkl")
pubmed_all # dim 111571 22
```

```python
# define dictionary
with open("../dictionary/psyc_unique_dict_847_2k-1k.txt") as f:
    psyc_dict = [x.strip('\n') for x in f.readlines()]
# psyc_dict = sorted(psyc_dict) # sort alphabetically
psyc_dict
```

```python
def countOccurrence(text, targetWords):
    frequency = Counter(text)
    targetFrequency = {word: frequency[word] for word in targetWords}
    return pd.Series(targetFrequency) # call Series s.t. the apply function returns a dataframe
```

```python
# filter data by time periods
# pubmed = pubmed_all.query("pubdate >= '1980' and pubdate <= '1989'") # 80-89
# pubmed = pubmed_all.query("pubdate >= '1990' and pubdate <= '1999'") # 90-99
# pubmed = pubmed_all.query("pubdate >= '2000' and pubdate <= '2009'") # 00-09
pubmed = pubmed_all.query("pubdate >= '2010' and pubdate <= '2015'") # 10-15
pubmed.shape, pubmed_all.shape
```

```python
# count occurrence (40 seconds / len=10) (1 min / len=172)
psyc_occurrence = pubmed['normalized_abstract'].apply(countOccurrence, targetWords=psyc_dict)
psyc_occurrence.insert(loc=0, column='pmid', value=pubmed.pmid)
psyc_occurrence
```

```python
word_freq = {}
for word in psyc_dict:
    word_freq[word] = sum(psyc_occurrence[word] > 0 ) / len(pubmed)
word_freq
```

```python
# word_freq_80to89 = word_freq.copy()
# word_freq_90to99 = word_freq.copy()
# word_freq_00to09 = word_freq.copy()
# word_freq_10to15 = word_freq.copy()
```

```python
word_freq_evolve = pd.DataFrame([word_freq_80to89, word_freq_90to99, 
                                 word_freq_00to09, word_freq_10to15]).T
word_freq_evolve.columns = ['80to89', '90to99', '00to09', '10to15']
word_freq_evolve = word_freq_evolve.reindex(psyc_dict) # descending overall word freq
word_freq_evolve
```
