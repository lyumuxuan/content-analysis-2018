```python
import glob
import pandas as pd
```

```python
filenames = glob.glob("./intermediate_data/*.csv")
print(filenames) # len 109
```

```python
combined_data = combined_csv.drop('Unnamed: 0', axis=1) # drop meaningless previous index
combined_data
```
